## Hacemos lo mismo pero iterando sobre todos los urls

```python
driver.implicitly_wait(10)
```

```python
vacia = []
# for url in urls[0:3]: 
for url in urls: 
    driver.get("https://www.tiendamonge.com/"+url)
    time.sleep(5)
    precios = driver.find_elements_by_class_name('prices')
    nombre = driver.find_elements_by_class_name('product-title')
    for i in range(len(nombre)):

        n = nombre[i].text
        p = precios[i].text
        vacia.append({'categoria': url , 'nombre': n, 'precio': p})
```

```python
categorias = pd.DataFrame(vacia)
categorias
```

```python
x = '₡ 289,900 ₡ 249,900\nOrden Especial'

def limpiar_precio(x):
    try:
        if len(x.replace('Orden Especial', '').replace('\n', '').replace(' ', '').split('₡')[-1].replace(',',''))!=0:
            return x.replace('Orden Especial', '').replace('\n', '').replace(' ', '').split('₡')[-1].replace(',','')
        else:
            return None

    except:
        return None

limpiar_precio(x)
```

```python
categorias['precio_limpio'] = categorias.precio.apply(limpiar_precio).astype(float)
```

```python
categorias[categorias.precio_limpio.isnull()]
```

```python
def sub_categoria(x):
    return x.split()[0].lower()
```

```python
categorias['sub_categoria'] = categorias.nombre.apply(sub_categoria)
```

```python
categorias
```

```python
marcas = ['HP','ACER','HUAWEI','MABE','FRIGIDAIRE','GENERAL ELECTRIC','TELSTAR','CICADEX','ALCATEL','MOTOROLA','LG','DELL','SONY' , 'SAMSUNG','PANASONIC','SERPENTO'
         , 'GO PRO', 'WHIRLPOOL', 'ATLAS', 'CAPRI', 'REMINGTON', 'CONAIR', 'WAHL', 'NIKON', 'SANDISK', 'BLACK & DECKER'
         , 'OSTER', 'LUXOR', 'STANLEY', 'TRUST', 'EPSON', 'COUGAR ARMOR', 'OSCAR CORRALES', 'PHILIPS', 'CAMA', 'COLCHON', 'MUEBLE'
         , 'COLCHÓN', 'DEFENDER', 'JUEGO DE SALA', 'SELE NB', 'EUROBIKE', 'SILLA', 'MESA']
```

```python
def marca(x):
    try:
        for i in marcas:
            if i in x:
                return i
    except:
        return 'otras'
```

```python
categorias['marca'] = categorias.nombre.apply(marca).fillna('otras')
```

```python
categorias.groupby(['marca']).size().sort_values(ascending=False)
```
